```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating data for the task
np.random.seed(42)  # For reproducibility
months = ['January', 'February', 'March', 'April', 'May', 'June']
product_a_sales = [100, 120, 130, 150, 160, 180]
product_b_sales = [80, 110, 140, 160, 190, 220]

# Combine into a DataFrame for convenience
data = pd.DataFrame({
    'Month': months,
    'Product_A_Sales': product_a_sales,
    'Product_B_Sales': product_b_sales
})
```

```python
data.shape
```

```python
data.head
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(data['Month'], data['Product_A_Sales'], marker='.', c="g", label="product A sales")
ax.plot(data['Month'], data['Product_B_Sales'], marker='.', c="y", label="product B sales")
ax.set_title("Product A vs. B sales")
ax.legend(loc="upper left")
ax.set_ylabel("Sales")
ax.set_xlabel("Months")
plt.show()
```
